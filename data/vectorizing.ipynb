```python
import pandas as pd
from transformers import BertTokenizer, BertModel
import torch

# Загрузка датасета
df = pd.read_csv("/content/drive/MyDrive/2019.csv")
```

```python
# Инициализация токенизатора и модели
tokenizer = BertTokenizer.from_pretrained('cl-tohoku/bert-base-japanese')
model = BertModel.from_pretrained('cl-tohoku/bert-base-japanese')
```

```python
# Функция для получения эмбеддингов
def get_embeddings(text):
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True)
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).numpy()[0]

# Применение функции к колонке `text`
df['embeddings'] = df['text'].apply(get_embeddings)
```

```python
# Сохранение DataFrame с эмбеддингами
df.to_csv('/content/drive/MyDrive/2019_with_embeddings.csv', index=False)

# Печать первых строк для проверки
print(df.head())
```
